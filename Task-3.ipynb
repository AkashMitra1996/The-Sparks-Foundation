# EDA on US Superstore Dataset

```python
#Importing libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
#Importing Dataset
data= pd.read_csv('SampleSuperstore.csv')
data.head()
```

```python
data.shape #shape of a given dataset
```

```python
data.columns #columns is dataset
```

```python
data.isna().sum() # check for null values or missing values
```

```python
data.info() #to check null values and variables type
```

```python
data.describe()  
```

```python
plt.figure(figsize= (15,9))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize= (15,9))
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
corr=data.corr() #correlation
corr
```

```python
#correlation matrix:
plt.figure(figsize=(7,6))
sns.heatmap(corr,annot=True,cmap='Blues')
```

```python
#DISTRIBUTION BASED ON REGION/CATEGORY
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.countplot('Region',data=data)
plt.title('DISTRIBUTION OF REGIONS')
plt.xlabel('REGION',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)


plt.subplot(1,2,2)
sns.countplot('Region',hue='Category',data=data)
plt.title('DISTRIBUTION OF REGIONS BASED ON CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISTRIBUTION BASED ON CATEGORY/SUBCATEGORY
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
sns.countplot('Category',data=data);
plt.title('DISTRIBUTION OF CATEGORY')
plt.xlabel('CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)

plt.subplot(2,1,2)
sns.countplot('Sub-Category',data=data);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot('Category',hue='Sub-Category',data=data);
plt.title('DISTRIBUTION OF SUB-CATEGORY')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
plt.show()
```

```python
#DISTRIBUTION of SEGMENTS BASED ON SHIPMENT MODE
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
sns.countplot('Segment',data=data)
plt.title('DISTRIBUTION OF SEGMENTS')
plt.xlabel('SEGMENT',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.grid(alpha=0.3)

plt.subplot(1,2,2)
sns.countplot('Ship Mode',data=data)
plt.title('DISTRIBUTION OF SHIP MODE')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('SHIP MODE',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=data)
plt.legend('Profit')
plt.grid(alpha=0.3)
```

```python
#DISTRIBUTION based on STATES
plt.figure(figsize=(30,10))
sns.countplot(data['State'])
plt.xticks(fontsize=16,rotation=90)
plt.ylabel('COUNTS',fontsize=20)
plt.xlabel('STATES',fontsize=20)
plt.title('DISTRIBUTION OF STATES',fontsize=25);
plt.grid(alpha=0.3)
plt.show()
```
